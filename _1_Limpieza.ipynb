```python
#Definimos las columnas y leimos el dataset
colnames = ['orden', 'armas_medios', 'barrios_hecho', 'latitud', 'longitud','zona','nom_comuna','año','mes','dia','dia_semana','descripcion_conducta','conducta','clasificaciones_delito','edad','curso_de_vida','estado_civil_persona','genero','movil_agresor','movil_victima']
df = pd.read_csv('delitos.csv', header=None,names=colnames, low_memory=False)
```

```python
#Ver 30 datos que necesitaremos preliminarmente
df[['año','dia','mes','clasificaciones_delito','barrios_hecho','latitud','longitud']].head(30)
```

```python
#Ver los tipos de datos del dataset
df.dtypes
```

```python
#Creamos un nuevo dataframe a partir del pasado para solo tener los datos que necesitaremos
df2=df[['año','dia','mes','clasificaciones_delito','conducta','armas_medios','barrios_hecho','nom_comuna','zona','latitud','longitud']]
```

```python
#Ver nuevo dataframe
df2
```

```python
#Cambiamos El tipo de datos de las columnas año, dia, longitud y latitud
df2['longitud'] = pd.to_numeric(df2['longitud'], errors='coerce')
df2['latitud'] = pd.to_numeric(df2['latitud'], errors='coerce')
df2['año'] = pd.to_numeric(df2['año'], errors='coerce')
df2['dia'] = pd.to_numeric(df2['dia'], errors='coerce')
```

```python
#Verificamos los cambios
df2.dtypes
```

```python
#Cambiamos el formato como estan escritos los meses para que no presenten un numero adelante
df2['mes'] = df['mes'].replace(['01. Enero','02. febrero','03. Marzo','04. Abril','05. Mayo','06. Junio','07. Julio','08. Agosto','09. Septiembre','10. Octubre','11. Noviembre','12. Diciembre'],['Enero','febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre'])
```

```python
#Verificamos el cambio
df2[['año','dia','mes','clasificaciones_delito','conducta','armas_medios','barrios_hecho','nom_comuna','zona','latitud','longitud']]
```

```python
#Verificamos cantidad de valores nulos en la columna latitud
df2['latitud'].isna().sum()
```

```python
#Verificamos cantidad de valores nulos en la columna longitud
df2['longitud'].isna().sum()
```

```python
#Eliminamos todas las filas que tengan un valor nulo
df2 = df2.dropna()
```

```python
#Verificamos el cambio
df2
```

```python
#Cambiamos tipo de datos para año y dia
df2['año'] = df2['año'].astype('int64', errors = 'ignore', copy=True,)
df2['dia'] = df2['dia'].astype('int64', errors = 'ignore', copy=True,)
```
